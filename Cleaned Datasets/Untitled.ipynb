```python
import pandas as pd
import streamlit as st
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec
```

```python
# load dataset
df_item = pd.read_csv("https://raw.githubusercontent.com/RYGcode/Brazilian-E-Commerce-Data-Analysis/main/Cleaned%20Datasets/cleaned_df_item.csv")
df_reviews = pd.read_csv("https://raw.githubusercontent.com/RYGcode/Brazilian-E-Commerce-Data-Analysis/main/Cleaned%20Datasets/cleaned_df_reviews.csv")
df_orders = pd.read_csv("https://raw.githubusercontent.com/RYGcode/Brazilian-E-Commerce-Data-Analysis/main/Cleaned%20Datasets/cleaned_df_orders.csv")
df_products = pd.read_csv("https://raw.githubusercontent.com/RYGcode/Brazilian-E-Commerce-Data-Analysis/main/Cleaned%20Datasets/cleaned_df_products.csv")
df_geolocation = pd.read_csv("https://raw.githubusercontent.com/RYGcode/Brazilian-E-Commerce-Data-Analysis/main/Cleaned%20Datasets/cleaned_df_geolocation.csv")
df_sellers = pd.read_csv("https://raw.githubusercontent.com/RYGcode/Brazilian-E-Commerce-Data-Analysis/main/Cleaned%20Datasets/cleaned_df_sellers.csv")
df_order_pay = pd.read_csv("https://raw.githubusercontent.com/RYGcode/Brazilian-E-Commerce-Data-Analysis/main/Cleaned%20Datasets/cleaned_df_order_pay.csv")
df_customers = pd.read_csv("https://raw.githubusercontent.com/RYGcode/Brazilian-E-Commerce-Data-Analysis/main/Cleaned%20Datasets/cleaned_df_customers.csv")
df_category = pd.read_csv("https://raw.githubusercontent.com/RYGcode/Brazilian-E-Commerce-Data-Analysis/main/Cleaned%20Datasets/cleaned_df_category.csv")
```

```python
# merging the datasets
df = pd.merge(df_orders,df_order_pay, on='order_id', how='inner')
df = df.merge(df_reviews, on='order_id', how='inner')
df = df.merge(df_customers, on='customer_id', how='inner')
df = df.merge(df_item, on='order_id', how='inner')
df = df.merge(df_products, on='product_id', how='inner')
df = df.merge(df_sellers, on='seller_id', how='inner')
df = df.merge(df_geolocation, left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='inner')
```
